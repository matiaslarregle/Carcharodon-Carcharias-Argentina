```python
# Importamos las librer√≠as necesarias
import pandas as pd
import folium
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
```

```python
# Cargamos los CSV en diferentes DataFrames
dfO = pd.read_csv('Occurrence.csv')
dfP = pd.read_csv('Pesca.csv')
dfB = pd.read_csv('BallenasBahia.csv')
dfDC = pd.read_csv('DistribucionCetaceos.csv')
dfEM = pd.read_csv('ElefantesMarinos.csv')
dfF = pd.read_csv('Focas.csv')

# Configuramos Pandas para que se muestren todas las columnas de los DF
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
pd.options.display.max_columns = None

dfO
```

```python
#Eliminamos todas las columnas que solo contienen valores nulos
dfO = dfO.dropna(axis=1, how='all')
dfP = dfP.dropna(axis=1, how='all')
dfB = dfB.dropna(axis=1, how='all')
dfDC = dfDC.dropna(axis=1, how='all')
dfEM = dfEM.dropna(axis=1, how='all')
dfF = dfF.dropna(axis=1, how='all')
```

```python
columna = dfO['class'].unique()

# Mostramos los diferentes valores de la columna 'class'
print("Valores distintos en la columna 'class':")
for valor in columna:
    print(valor)
```

```python
# Creamos un nuevo DF tomando dos valores en una columna del original
dfO_filtrado = dfO[dfO['class'].isin(['Elasmobranchii', 'Teleostei'])]
```

```python
columna = dfO['order'].unique()

print("Valores distintos en la columna 'order':")
for valor in columna:
    print(valor)
```

```python
selachii = dfO[dfO['infraclass'] == 'Selachii']
selachii
```

```python
lamniformes = selachii[selachii['order'] == 'Lamniformes']
lamniformes
```

```python
lamnidae = lamniformes[lamniformes['family'] == 'Lamnidae']
lamnidae
```

```python
columna = dfDC['genus'].unique()

print("Valores distintos en la columna 'genus':")
for valor in columna:
    print(valor)
```

```python
valor1 = 'Orcinus orca'
valor2 = 'Seudorca crassidens'

# Filtramos el DF eliminando las filas que contienen ciertos valores en una determinada columna
dfDC_filtrado = dfDC[(dfDC['scientificname'] != valor1) & (dfDC['scientificname'] != valor2)]
```

```python
# Creamos otro DF que solamente contiene dos especies
Orcas = dfDC[(dfDC['scientificname'] == 'Orcinus orca') | (dfDC['scientificname'] == 'Seudorca crassidens')]
```

```python
# Creamos un mapa interactivo con Plotly Express que muestre las ubicaciones de las orcas
fig = px.scatter_geo(Orcas, lat='decimallatitude', lon='decimallongitude', hover_name='scientificname', scope='world')

# Mostramos el mapa
fig.show()
```

```python
fig = px.scatter_geo(dfO, lat='decimallatitude', lon='decimallongitude', hover_name='scientificname', scope='world')

fig.show()
```

```python
# Concaatenamos DF que contienen informaci√≥n de pinn√≠pedos
pinnipedos = pd.concat([dfEM, dfF], ignore_index=True)
```

```python
# Creamos un DF concatenando todos
df_combined = pd.concat([pinnipedos, dfO_filtrado, Orcas, dfDC_filtrado, lamnidae], ignore_index=True)

# Creamos una nueva columna para calcular la cantidad total de registros de especies en cada ubicaci√≥n
df_combined['Total_Registros'] = df_combined.groupby(['decimallatitude', 'decimallongitude'])['scientificname'].transform('count')
```

```python
# Creamos un mapa que grafica seg√∫n la concentraci√≥n de registros en cada punto
fig = px.scatter_geo(df_combined, 
                     lat='decimallatitude', 
                     lon='decimallongitude', 
                     color='Total_Registros',  # Asignamos una columna a tomar en cuenta para graficar seg√∫n su concentraci√≥n 
                     color_continuous_scale='Viridis', # Escala de colores
                     hover_name='scientificname',  # Informaci√≥n adicional para cada punto
                     title='Mapa de ubicaciones √≥ptimas para el Tibur√≥n Blanco')  # Titulamos el mapa

fig.show()
```

```python
grid_size = 1  # Asignamos un tama√±o a la grilla que divide el mapa

lon_min, lon_max = df_combined['decimallongitude'].min(), df_combined['decimallongitude'].max()
lat_min, lat_max = df_combined['decimallatitude'].min(), df_combined['decimallatitude'].max()
lon_range = np.arange(lon_min, lon_max + grid_size, grid_size)
lat_range = np.arange(lat_min, lat_max + grid_size, grid_size)

# Calculamos la puntuaci√≥n promedio en cada celda de la grilla
heatmap_data = []
for lat in lat_range:
    for lon in lon_range:
        # Filtramos los datos dentro de la celda de la grilla
        filtered_data = df_combined[
            (df_combined['decimallatitude'] >= lat) & (df_combined['decimallatitude'] < lat + grid_size) &
            (df_combined['decimallongitude'] >= lon) & (df_combined['decimallongitude'] < lon + grid_size)
        ]
        # Calcular la puntuaci√≥n promedio para esta celda
        if len(filtered_data) > 0:
            avg_score = filtered_data['Total_Registros'].mean()
            
            # Ajustamos la puntuaci√≥n mediante criterios espec√≠ficos:
            
            # Aumentamos el peso si hay registros de cetaceos o pinn√≠pedos
            if filtered_data['infraorder'].isin(['Cetacea', 'Pinnipedia']).any():
                avg_score *= 1.7 
            
            # Mantenemos el peso si hay avistamientos de elasmobranquios o tele√≥steos
            if filtered_data['class'].isin(['Elasmobranchii', 'Teleostei']).any():
                avg_score *= 1  
            
            # Reducimos el peso si hay avistamientos de orcas
            if 'Orcinus orca' in filtered_data['scientificname'].values:
                avg_score *= 0.5  
            
            # Aumentamos el peso si hay avistamientos de l√°mnidos
            if filtered_data['family'].eq('Lamnidae').any():
                avg_score *= 4  
                
            # Ajustamos la puntuaci√≥n seg√∫n la ubicaci√≥n geogr√°fica
            
            # Reducimos el puntaje al sur de la Bah√≠a Camarones
            if lat < -44:  
                avg_score *= 0.25  
            # Anulamos los valores del Pac√≠fico
            if lon < -70:  
                avg_score *= 0  
                
            heatmap_data.append({'lat': lat + grid_size / 2, 'lon': lon + grid_size / 2, 'score': avg_score})

fig = px.density_mapbox(heatmap_data, 
                        lat='lat', 
                        lon='lon', 
                        z='score',
                        radius=25, 
                        opacity=0.5, 
                        center=dict(lat=-38.5769, lon=-58.6935,), # Definimos el centro de nuestro mapa
                        zoom=2.5,
                        mapbox_style='open-street-map')

# Agregamos puntos en ubicaciones espec√≠ficas

fig.add_trace(go.Scattermapbox(
    lat=[-38.5769, -43.3185, -38.2728],
    lon=[-58.6935, -65.0440, -57.8284],
    marker=go.scattermapbox.Marker(
        size=5,
        color='rgb(255, 0, 0)',
        opacity=0.7
    ),
    text=['Puerto Quequ√©n', 'Playa Uni√≥n', 'Miramar'],
    hoverinfo='text'
))

# Ajustamos el marco
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})

fig.show()
```
